```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('musics.csv')
```

```python
data.dropna(inplace=True)
```

```python
data=data.drop_duplicates()
```

```python
l=[]
for i in data['User-Rating']:
  l.append(i[:3])
l
```

```python
data['User-Rating']=l
data
```

```python
data.info()
```

```python
data['Album/Movie']=data['Album/Movie'].str.replace(' ','')
data['Singer/Artists']=data['Singer/Artists'].str.replace(' ','')
```

```python
data['Singer/Artists']=data['Singer/Artists'].str.replace(',',' ')
```

```python
data['tags']=data['Album/Movie']+' '+data['Singer/Artists']+' '+data['Genre']+' '+data['User-Rating']
```

```python
data
```

```python
df=data[['Song-Name','tags']]
df
```

```python
df['tags']=df['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2000,stop_words='english')
```

```python
vectors=cv.fit_transform(df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
df.rename(columns={'Song-Name':'SongName'},inplace=True)
```

```python
def recommend(music):
  music_index=df[df['SongName']==music].index[0]
  distances=similarity[music_index]
  music_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in music_list:
    print(df.iloc[i[0]].SongName)
```

```python
recommend('Coca Cola')
```
